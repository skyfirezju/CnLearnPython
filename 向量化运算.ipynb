```python
import pandas
data = pandas.DataFrame(
    data = {
        'A' : [1, 3, 5, 7, 9]
    }
)
data
```

```python
data['B'] = data.A.pow(2)
data
```

```python
# 加法
data['add'] = data.B + data.A
# 减法
data['sub'] = data.B - data.A
# 乘法
data['mul'] = data.B * data.A
# 除法
data['div'] = data.B / data.A
data
```

```python
for a in data['A']:
    print(a)
```

```python
C = []
for a in data['A']:
    C.append(a*a*a)
data['C'] = C
data
```

```python
for index, r in data.iterrows():
    print(index)
    print(r)
    print('--------')
```

```python
D = []
for index, r in data.iterrows():
    # 把 A+B 的结果追加到 D中
    D.append(r['A'] + r['B'])
data['D'] = D
data
```
