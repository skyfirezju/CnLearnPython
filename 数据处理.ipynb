```python
import pandas
items = pandas.read_csv(
    '/Users/xiefei/Documents/Workspace/CnLearnPython/PDABook/第四章/4.5.3 字段匹配/商品名称.csv',
    engine = 'python'
)
prices = pandas.read_csv(
    '/Users/xiefei/Documents/Workspace/CnLearnPython/PDABook/第四章/4.5.3 字段匹配/商品价格.csv',
    engine = 'python'
)
```

```python
items
```

```python
prices
```

```python
# 默认内连接
itemPrices = pandas.merge(
    items,
    prices,
    left_on = 'id',
    right_on = 'id'
)
itemPrices
```

```python
# 左连接
itemPrices = pandas.merge(
    items,
    prices,
    left_on = 'id',
    right_on = 'id',
    how = 'left'
)
itemPrices
```

```python
# 右连接
itemPrices = pandas.merge(
    items,
    prices,
    left_on = 'id',
    right_on = 'id',
    how = 'right'
)
itemPrices
```

```python
# 外连接
itemPrices = pandas.merge(
    items,
    prices,
    left_on = 'id',
    right_on = 'id',
    how = 'outer'
)
itemPrices
```

```python
import pandas
data = pandas.read_csv(
    '/Users/xiefei/Documents/Workspace/CnLearnPython/PDABook/第四章/4.6.2 时间计算/时间计算.csv',
    engine = 'python'
)
data
```
